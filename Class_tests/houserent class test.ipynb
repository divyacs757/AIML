```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('D:/dip(AIML(cse))/kaggle/House_Rent_Dataset.csv')
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
mean=df['Size'].mean()
df['Size']=df['Size'].fillna(mean)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df.index,df['Rent'])
```

```python
q1=df['Rent'].quantile(0.25)
q3=df['Rent'].quantile(0.75)
iqr=q3-q1
upper_bound=q3+1.5*iqr
lower_bound=q1-1.5*iqr
```

```python
outer=df[(df['Rent']>=lower_bound)&(df['Rent']<=upper_bound)]
plt.scatter(outer.index,outer['Rent'].values)
```

```python
count=df['BHK'].value_counts()
plt.figure(figure=(6,6))
plt.pie(countlabels=count.index,autopct='1%.1f%%')
```
