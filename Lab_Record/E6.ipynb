## 1.Consider student dataset & train simple linear regression to predict student scores based on study hours and also Evaluate the model using different metrics.


```python
# import packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pd.read_csv('D:/dip(AIML(cse))/kaggle/StudentStudyHour1.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
# dropping rows which have null values
df.dropna(inplace=True,axis=0)
df.head(5)
```

```python
df.isnull().sum()
```

```python
#splits the dataset into dependent(y) and independent(x) variables
Y = df['Scores']
X = df.drop('Scores',axis=1)
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.25)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
print(x_train)  #just to check the values in test data
```

```python
print(y_test) #just to check the values in test data
```

```python
# train the model using training data
lr = LinearRegression()
model = lr.fit(x_train,y_train)
```

```python
#test the model using test data ( Pass only independent variable not dependent varible)
y_pred = model.predict(x_test)

# Just compare actual values and predicted values
df = pd.DataFrame({'y_test':y_test,'y_pred':y_pred})
df
```

```python
# Evaluate the model using MSE RMSE
print("The MSE is : ", mean_squared_error(y_test,y_pred))
print("The R square is : ", r2_score(y_test,y_pred))
```
