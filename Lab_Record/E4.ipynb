### Univariate and Bivariate Notes

### Define Univariate
#### Univariate data refers to a dataset where each observation is associated 
#### with only one variable. 
#### This means it focuses on measuring or observing a single characteristic 
#### or attribute for each individual in the dataset.



```python
#import libries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('D:/dip(AIML(cse))/kaggle/auto-mpg.csv')
df
```

# Histogram
## cylinders counts

```python
df['cylinders'].plot(kind='hist')
plt.title('counting cylinders')
plt.xlabel('cylinders')
plt.ylabel('cylinders counts')
```

## line graph
### Number of Cars by Model Year

```python
df['model year'].value_counts().sort_index().plot(kind='line')
plt.xlabel('Model Year')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Model Year')
plt.show()
```

 ## ploting boxplot
 ### MPG Distribution by weight

```python
plt.boxplot(df['weight'])
plt.title('MPG Distribution by weight')
plt.ylabel('weight')
plt.show()
```

 ## bar graph
 ### count of cylinders 

```python
df['cylinders'].value_counts().plot(kind='bar')
```

## pie chart
### count of origins

```python
df['origin'].value_counts().plot(kind='pie',autopct='%1.2f%%')
plt.title('count origins')
```

## Using titanic dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('D:/dip(AIML(cse))/kaggle/titanic.csv')
df
```

## line graph
### plot line graph to visualize the comparison of male and female survival counts.

```python
a=df.groupby(['survived','sex']).size().unstack()
a.plot(kind='line',marker='*')
```

## bar plot
### Count the number of passengers in each group and plot a bar graph to compare the survival and non-survival counts of males and females

```python
a=df.groupby(['sex','survived']).size().unstack()
a.plot(kind='bar')
```

## scatter plot
### plot a scatter plot to analyze the relationship between passenger age (age) and ticket fare (fare).

```python
plt.scatter(df['age'], df['fare'])
plt.xlabel('Age')
plt.ylabel('fare')
plt.title('fare vs Age')
plt.show()
```
