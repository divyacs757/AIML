```python
# Import libraries
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Step 1: Define text directly
test = "Hello everyone! How is the class today? The hotel was very clean, and the food was amazing. I loved the staff and service."
```

```python
# Step 2: Sentence Tokenization
sentences = sent_tokenize(test)
print("Sentence Tokenization:")
print(sentences)
```

```python
# Step 3: Word Tokenization
words = word_tokenize(test)
print("Word Tokenization:")
print(words)
```

```python
# Step 4: Text Cleaning (remove punctuation and lowercase)
words = [word.lower() for word in words if word.isalpha()]
print(words)
```

```python
# Step 5: Remove Stopwords
filtered_words = [word for word in words if word not in stopwords.words('english')]
print(filtered_words)
```

```python
# Step 6: Frequency Distribution - Bar Chart
fdist = FreqDist(filtered_words)
w, c = zip(*fdist.most_common(10))
plt.bar(w, c)
plt.title("Top 10 Words")
plt.show()
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_words))
plt.imshow(wordcloud, interpolation='bilinear')
```
