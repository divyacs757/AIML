# Consider student dataset & train Multiple Linear Regression to predict boston house and  also Evaluate the model using different metrics.

```python
# import packages
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv('D:/dip(AIML(cse))/kaggle/Boston.csv')
df.head()
```

```python
# check for missing values in all the columns
df.isnull().sum()
```

```python
mean=df['chas'].mean()
df['chas']=df['chas'].fillna(mean)
df.isnull().sum()
```

```python
X = df.drop('medv',axis=1)
Y = df['medv'] #This "medv" is taken as target price to predict
```

```python
from sklearn.model_selection import train_test_split

# splits the training and test data set in 80% : 20%
# assign random_state to any value.This ensures consistency.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
#test the model using test data ( Pass only independent variable not dependent varible)
y_pred = lin_model.predict(X_test)

# Just compare actual values and predicted values
df = pd.DataFrame({'Y_test':Y_test,'y_pred':y_pred})
df
```

```python
# Evaluate the model using MSE RMSE
print("The MSE is : ", mean_squared_error(Y_test,y_pred))
print("The R square is : ", r2_score(Y_test,y_pred))
```
