## 1.Consider breast_cancer dataset and perform the following operations
### a)build a model using Decision tree algorithm
### b)Test the model using different metrics
### c)Compare the decision tree model with Random Forest Algorith

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('D:/dip(AIML(cse))/kaggle/Breast_Cancer_data.csv')
df.head()
```

```python
x=df.drop('diagnosis',axis=1)
y=df['diagnosis']
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2,random_state=101)
x_test.shape,x_train.shape,y_test.shape,y_train.shape
```

## a)build a model using Decision tree algorithm

```python
dtc=DecisionTreeClassifier()
model=dtc.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

## b)Test the model using different metrics

```python
# Accuracy:
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, x_test, y_test, cmap=plt.cm.Reds)
plt.show()
```

```python
#different metrics 
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## c)Compare the decision tree model with Random Forest Algorithm

```python
rfc=RandomForestClassifier()
model1=rfc.fit(x_train,y_train)
y_pred=model1.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model1, x_test, y_test, cmap=plt.cm.Reds)
plt.show()
```
