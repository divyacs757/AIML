## ->Read the Titanic Dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('D:/dip(AIML(cse))/kaggle/titanic.csv')
df
```

## Examples:
### 1.Categorical Data
### A bar graph is a visual representation of data using rectangular bars to show the values of different categories. The length or height of each bar corresponds to the data value it represents.


#### ->plot a bar chart that shows how many males and females were on board. Label the axes and give the plot the title “Count of Sex”

```python
#bor graph
a= df['sex'].value_counts().sort_index()
plt.bar(a.index, a.values)
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('count of sex')
plt.show()
```

#### ->plot a bar chart to compare the ages of passengers with respect to their survival status (alive)

```python
#bar graph
plt.bar(df['alive'],df['age'])
plt.title("alive and not alive passengers")
plt.xlabel("alive")
plt.ylabel("passengers age")
plt.show()
```

### 2.Distribution
### A histogram is a graphical representation that shows the distribution of a numerical dataset using bars. Each bar represents the frequency of data points within a specific range (bin).


#### ->plot a histogram to show the distribution of passenger ages. Divide the ages into 10 equal groups (bins)

```python
plt.hist(df['age'],bins=10)
plt.title("distrubution of age")
plt.xlabel("group of age")
plt.ylabel("passengers")
plt.show()
```


### A scatter plot is a graph that displays points to show the relationship between two different variables. Each point represents an observation with values on the x- and y-axes.

#### ->plot a scatter plot to study the relationship between passenger age (age) and ticket fare (fare).





```python
plt.scatter(df['age'],df['fare'])
plt.title("distrubution of age vs fare")
plt.xlabel("group of age")
plt.ylabel("fare")
plt.show()
```

### 3.Trends
### A line graph is a chart that connects data points with a continuous line to show trends over time or ordered categories. It’s useful for visualizing changes and patterns in data.








```python
a=df['class'].value_counts()
a
```

#### ->plot a line graph to visualize the number of passengers in each class.

```python
a=df['class'].value_counts()
plt.title("count of class")
plt.xlabel("class")
plt.ylabel("passengers")
a.plot(kind='line')
```

### 4.Proportions
### A pie chart is a circular graph divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole.






## ->plot a pie chart to show the percentage of passengers in each travel class (class).

```python
plt.pie(df['class'].value_counts(),labels=[1,2,3],autopct='%1.2f%%')
plt.show()
```

### 5.Uncertainties / Spread
### A boxplot is a graphical display that shows the distribution of a dataset through its quartiles. It highlights the median, spread, and potential outliers in the data.



#### ->draw a box plot to visualize the distribution of ticket fares (fare)

```python
plt.boxplot(df['fare'])
plt.title("distribution of fare")
plt.ylabel("fare")
plt.show()
```
