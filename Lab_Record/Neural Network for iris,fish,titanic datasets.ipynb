# Iris

```python
import pandas as pd 
import numpy as np 
from keras.layers import Dense 
from keras.models import Sequential 
from keras.optimizers import Adam 
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("D:/dip(AIML(cse))/kaggle/iris_flower.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
m=df['SepalWidthCm'].mean()
df['SepalWidthCm']=df['SepalWidthCm'].fillna(m)
df.isnull().sum()
```

```python
x=df.drop(['Id','Species'],axis=1)
y=df['Species']
```

```python
y=LabelEncoder().fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=789)
x_train.shape[1]
```

```python
model=Sequential()
model.add(Dense(10,input_shape=(4,),activation='relu',name='input'))
model.add(Dense(10,activation='relu',name='fa2'))
model.add(Dense(3,activation='sigmoid',name='output'))
```

```python
optimizer=Adam(learning_rate=0.001)
model.compile(optimizer,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=8,verbose=1)
```

```python
model.evaluate(x_test,y_test)
```

# Titanic

```python
df=pd.read_csv("D:/dip(AIML(cse))/kaggle/titanic.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
m=df['age'].mean()
df['age']=df['age'].fillna(m)
```

```python

df['embarked']=df['embarked'].fillna('S')
```

```python

df['deck']=df['deck'].fillna('C')
```

```python

df['embark_town']=df['embark_town'].fillna('Southampton')
```

```python
df.isnull().sum()
```

```python
x=df.drop('survived',axis=1)
y=df['survived']
```

```python
x=pd.get_dummies(x)
y=LabelEncoder().fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
x_train.shape[1]
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model = Sequential()
model.add(Dense(10,input_shape=(30,),activation='relu',name='input'))
model.add(Dense(10,activation='relu',name='fc2')) 
model.add(Dense(2,activation='softmax',name='output')) 
```

```python
optimizer=Adam(learning_rate=0.001)
model.compile(optimizer,loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print('NNM Summary:')
print(model.summary())
```

```python
model.fit(x_train, y_train,epochs=10, batch_size=8, verbose=1)
model.evaluate(x_test,y_test)
```
